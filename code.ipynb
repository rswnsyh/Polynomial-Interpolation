```
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

#Lagrange function

def approx_lagrange(x, y, a):

  X = sym.Symbol("X")   
  n = len(x)            
  P = 0                 
  
  for i in range (n):
    atas = 1            
    bawah = 1           
    for j in range (n):
      if i != j :
        atas = atas*(X-x[j])        
        bawah = bawah*(x[i]-x[j])   
    L = atas/bawah
    P = P + y[i]*L      
  
  f = sym.expand(P)     
  
  f_a = f.subs(X,a)     
  
  return f_a
```

```
#Testing the code using Sin function

x1 = np.arange(0, 5.25, 0.25)
y1 = np.sin(x1)

a = [0.5 , 1.5 , 2.5 , 3.4 , 4.5]

fa = np.zeros(len(a))     
for i in range (len(a)):
  fa[i] = approx_lagrange([0, 1, 2, 3, 4, 5], [np.sin(0), np.sin(1), np.sin(2), np.sin(3), np.sin(4), np.sin(5)], a[i])
  
plt.plot(x1, y1)
plt.plot(a, fa, 'o')
plt.legend(('Polynomial', 'Approximation'))

plt.show()
```
